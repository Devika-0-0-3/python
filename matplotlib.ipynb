```python
from numpy import array
from numpy.linalg import inv
a=array([[1.0,2.0],[3.0,4.0]])
print(a)
b=inv(a)
print(b)
i=a.dot(b)
print(i)
```

```python
import numpy as np
a=np.array([[1.0,2.0],[3.0,4.0]])
print(a)
b=np.linalg.inv(a)
print(b)
i=a.dot(b)
print(i)
```

```python
#adding grid and legen to the plot
from pylab import *
x=np.arange(0,math.pi*2,0.05)
plot(x,sin(x),label="sin")
plot(x,cos(x),"r",label="cos")
plot(x,-sin(x),"g--",label="-sin")
xlabel("x")
ylabel("sin cos -sin")
title("Waves")
grid(True)
legend(loc="center")
show()
```

```python
#subplot within same plot
from pylab import *

x=np.arange(0,math.pi*2,0.05)
#splot1
subplot(2,2,1)
plot(x,sin(x),label="sin")
legend(loc="upper right")
xlabel("x")
ylabel("sin")
grid(True)
#splot2
subplot(2,2,2)
grid(True)
plot(x,cos(x),label="cos")
xlabel("x")
ylabel("cos")
legend(loc="upper right")
#splot3
subplot(2,2,3)
xlabel("x")
ylabel("-sin")
grid(True)
plot(x,-sin(x),label="-sin")
legend(loc="upper right")
#splot4
subplot(2,2,4)
xlabel("x")
ylabel("tan")
grid(True)
plot(x,tan(x),label="tan")
legend(loc="upper right")


show()



```
